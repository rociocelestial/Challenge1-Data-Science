### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
suma_total_precio = tienda['Precio'].sum()
suma_total_precio2 = tienda2['Precio'].sum()
suma_total_precio3 = tienda3['Precio'].sum()
suma_total_precio4 = tienda4['Precio'].sum()

print(f'El valor total de la tienda 1 es: ${suma_total_precio}')
print(f'El valor total de la tienda 2 es: ${suma_total_precio2}')
print(f'El valor total de la tienda 3 es: ${suma_total_precio3}')
print(f'El valor total de la tienda 4 es: ${suma_total_precio4}')
```

```python
import matplotlib.pyplot as plt

tiendas =['Tienda', 'Tienda2','Tienda3','Tienda4']
volores_totales = [suma_total_precio,suma_total_precio2,suma_total_precio3,suma_total_precio4]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, volores_totales, color=['blue','green','red','purple'])
plt.title('Ventas totales por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ventas totales($)')
plt.show()

```

# 2. Ventas por categoría

```python
vpc_tienda1 = tienda['Categoría del Producto'].value_counts()
vpc_tienda2 = tienda2['Categoría del Producto'].value_counts()
vpc_tienda3 = tienda3['Categoría del Producto'].value_counts()
vpc_tienda4 = tienda4['Categoría del Producto'].value_counts()

print(f'Ventas por categoría de la tienda 1: \n{vpc_tienda1}')
```

```python
print(f'Ventas por categoría de la tienda 2: \n{vpc_tienda1}')
```

```python
print(f'Ventas por categoría de la tienda 3: \n{vpc_tienda1}')
```

```python
print(f'Ventas por categoría de la tienda 4: \n{vpc_tienda1}')
```

```python
venta_por_categoria = tienda.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)
venta_por_categoria2 = tienda2.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)
venta_por_categoria3 = tienda3.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)
venta_por_categoria4 = tienda4.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)

print(f'Ventas por categoría de la tienda 1: \n{venta_por_categoria}')
```

```python
print(f'Ventas por categoría de la tienda 2: \n{venta_por_categoria2}')
```

```python
print(f'Ventas por categoría de la tienda 3: \n{venta_por_categoria3}')
```

```python
print(f'Ventas por categoría de la tienda 4: \n{venta_por_categoria4}')
```

```python
df_ventas = pd.DataFrame({
    'Tienda': venta_por_categoria['Precio'],
    'Tienda2': venta_por_categoria2['Precio'],
    'Tienda3': venta_por_categoria3['Precio'],
    'Tienda4': venta_por_categoria4['Precio']
})

plt.figure(figsize=(12, 7))
df_ventas.plot(kind='bar', ax=plt.gca())
plt.title('Magnitud de Ventas Totales por categoría en Cuatro Tiendas ')
plt.xlabel('Categoría del Producto')
plt.ylabel('Ventas por Categoria($)')
plt.legend(title='Tienda',loc='upper right')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio = tienda['Costo de envío'].mean()
costo_envio_promedio2 = tienda2['Costo de envío'].mean()
costo_envio_promedio3 = tienda3['Costo de envío'].mean()
costo_envio_promedio4 = tienda4['Costo de envío'].mean()

print(f'El costo de envío  promedio de la tienda 1 es de: {costo_envio_promedio:.2f}')
print(f'El costo de envío  promedio de la tienda 2 es de: {costo_envio_promedio2:.2f}')
print(f'El costo de envío  promedio de la tienda 3 es de: {costo_envio_promedio3:.2f}')
print(f'El costo de envío  promedio de la tienda 4 es de: {costo_envio_promedio4:.2f}')
```

```python
tiendas =['Tienda', 'Tienda2','Tienda3','Tienda4']
volores_totales_costo_promedio= [costo_envio_promedio,costo_envio_promedio2,costo_envio_promedio3,costo_envio_promedio4]

plt.figure(figsize=(10, 6))
plt.pie(volores_totales_costo_promedio, labels=tiendas, autopct='%1.1f%%', colors=['blue','green','red','purple'])
plt.title('Distribución del Costo de Envío Promedio por Tienda')
plt.axis('equal') # Para asegurar que el gráfico circular sea un círculo.
plt.show()
```

# 6. Visualización

```python
import matplotlib.pyplot as plt
```

```python
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

df_geo = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
df_geo[['Tienda','lat','lon']].head()
```

```python
plt.figure(figsize=(9, 7))

# Obtener los nombres de las tiendas únicas
unique_tiendas = df_geo['Tienda'].unique()

# Definir una lista de colores (o usar un mapa de colores de matplotlib)
colors = ['green', 'yellow', 'skyblue', 'pink'] # Se corrigió 'sky blue' a 'skyblue'

# Iterar sobre cada tienda para graficar sus puntos
for i, tienda_name in enumerate(unique_tiendas):
    df_subset = df_geo[df_geo['Tienda'] == tienda_name]
    plt.scatter(
        data=df_subset,
        x='lon',
        y='lat',
        s=40,
        label=tienda_name, # Etiqueta para la leyenda
        color=colors[i % len(colors)], # Asignar un color de la lista
        alpha=0.6 # Hacer los puntos semi-transparentes
    )

plt.title('Distribución Geográfica de las Ventas por Tiendas')
plt.xlabel('Longitud (lon)')
plt.ylabel('Latitud (lat)')
plt.legend(title='Tienda')
plt.show()
```

# 7. Fecha Promedio de Compra por Tienda

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convertir la columna 'Fecha de Compra' a tipo datetime antes de calcular la media
tienda['Fecha de Compra'] = pd.to_datetime(tienda['Fecha de Compra'], format='%d/%m/%Y')
tienda2['Fecha de Compra'] = pd.to_datetime(tienda2['Fecha de Compra'], format='%d/%m/%Y')
tienda3['Fecha de Compra'] = pd.to_datetime(tienda3['Fecha de Compra'], format='%d/%m/%Y')
tienda4['Fecha de Compra'] = pd.to_datetime(tienda4['Fecha de Compra'], format='%d/%m/%Y')

fecha_compra_promedio = tienda['Fecha de Compra'].mean()
fecha_compra_promedio2 = tienda2['Fecha de Compra'].mean()
fecha_compra_promedio3 = tienda3['Fecha de Compra'].mean()
fecha_compra_promedio4 = tienda4['Fecha de Compra'].mean()

print(f'La fecha promedio de compra de la tienda 1 es de: {fecha_compra_promedio.strftime('%Y-%m-%d')}')
print(f'La fecha promedio de compra de la tienda 2 es de: {fecha_compra_promedio2.strftime('%Y-%m-%d')}')
print(f'La fecha promedio de compra de la tienda 3 es de: {fecha_compra_promedio3.strftime('%Y-%m-%d')}')
print(f'La fecha promedio de compra de la tienda 4 es de: {fecha_compra_promedio4.strftime('%Y-%m-%d')}')





```
