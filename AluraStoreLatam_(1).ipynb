### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
suma_total_precio = tienda['Precio'].sum()
suma_total_precio2 = tienda2['Precio'].sum()
suma_total_precio3 = tienda3['Precio'].sum()
suma_total_precio4 = tienda4['Precio'].sum()

print(f'El valor total de la tienda 1 es: ${suma_total_precio}')
print(f'El valor total de la tienda 2 es: ${suma_total_precio2}')
print(f'El valor total de la tienda 3 es: ${suma_total_precio3}')
print(f'El valor total de la tienda 4 es: ${suma_total_precio4}')
```

# 2. Ventas por categoría

```python
vpc_tienda1 = tienda['Categoría del Producto'].value_counts()
vpc_tienda2 = tienda2['Categoría del Producto'].value_counts()
vpc_tienda3 = tienda3['Categoría del Producto'].value_counts()
vpc_tienda4 = tienda4['Categoría del Producto'].value_counts()

print(f'Ventas por categoría de la tienda 1: \n{vpc_tienda1}')
```

```python
print(f'Ventas por categoría de la tienda 2: \n{vpc_tienda1}')
```

```python
print(f'Ventas por categoría de la tienda 3: \n{vpc_tienda1}')
```

```python
print(f'Ventas por categoría de la tienda 4: \n{vpc_tienda1}')
```

```python
venta_por_categoria = tienda.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)
venta_por_categoria2 = tienda2.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)
venta_por_categoria3 = tienda3.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)
venta_por_categoria4 = tienda4.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)

print(f'Ventas por categoría de la tienda 1: \n{venta_por_categoria}')
```

```python
print(f'Ventas por categoría de la tienda 2: \n{venta_por_categoria2}')
```

```python
print(f'Ventas por categoría de la tienda 3: \n{venta_por_categoria3}')
```

```python
print(f'Ventas por categoría de la tienda 4: \n{venta_por_categoria4}')
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio = tienda['Calificación'].mean()
calificacion_promedio2 = tienda2['Calificación'].mean()
calificacion_promedio3 = tienda3['Calificación'].mean()
calificacion_promedio4 = tienda4['Calificación'].mean()

print(f'La calificación promedio de la tienda 1 es de: {calificacion_promedio:.2f}')
```

```python
print(f'La calificación promedio de la tienda 2 es de: {calificacion_promedio2:.2f}')
```

```python
print(f'La calificación promedio de la tienda 3 es de: {calificacion_promedio3:.2f}')
```

```python
print(f'La calificación promedio de la tienda 4 es de: {calificacion_promedio4:.2f}')
```
