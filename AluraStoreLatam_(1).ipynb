### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
suma_total_precio = tienda['Precio'].sum()
suma_total_precio2 = tienda2['Precio'].sum()
suma_total_precio3 = tienda3['Precio'].sum()
suma_total_precio4 = tienda4['Precio'].sum()

print(f'El valor total de la tienda 1 es: ${suma_total_precio}')
print(f'El valor total de la tienda 2 es: ${suma_total_precio2}')
print(f'El valor total de la tienda 3 es: ${suma_total_precio3}')
print(f'El valor total de la tienda 4 es: ${suma_total_precio4}')
```

```python
import matplotlib.pyplot as plt

tiendas =['Tienda', 'Tienda2','Tienda3','Tienda4']
volores_totales = [suma_total_precio,suma_total_precio2,suma_total_precio3,suma_total_precio4]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, volores_totales, color=['blue','green','red','purple'])
plt.title('Ventas totales por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ventas totales($)')
plt.show()

```

```python
venta_por_categoria = tienda.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)
venta_por_categoria2 = tienda2.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)
venta_por_categoria3 = tienda3.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)
venta_por_categoria4 = tienda4.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)

print(f'Ventas por categoría de la tienda 1: \n{venta_por_categoria}')
```

```python
df_ventas = pd.DataFrame({
    'Tienda': venta_por_categoria['Precio'],
    'Tienda2': venta_por_categoria2['Precio'],
    'Tienda3': venta_por_categoria3['Precio'],
    'Tienda4': venta_por_categoria4['Precio']
})

plt.figure(figsize=(12, 7))
df_ventas.plot(kind='bar', ax=plt.gca())
plt.title('Magnitud de Ventas Totales por categoría en Cuatro Tiendas ')
plt.xlabel('Categoría del Producto')
plt.ylabel('Ventas por Categoria($)')
plt.legend(title='Tienda',loc='upper right')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Producto más vendido (nombre)
ventas_producto_tienda1 = tienda['Producto'].value_counts().idxmax()
ventas_producto_tienda2 = tienda2['Producto'].value_counts().idxmax()
ventas_producto_tienda3 = tienda3['Producto'].value_counts().idxmax()
ventas_producto_tienda4 = tienda4['Producto'].value_counts().idxmax()

# Cantidad maxima del producto vendido
maxima_cantidad_vendida_tienda1 = tienda['Producto'].value_counts().max()
maxima_cantidad_vendida_tienda2 = tienda2['Producto'].value_counts().max()
maxima_cantidad_vendida_tienda3 = tienda3['Producto'].value_counts().max()
maxima_cantidad_vendida_tienda4 = tienda4['Producto'].value_counts().max()

print(f'El producto más vendido en la tienda 1 es {ventas_producto_tienda1} con {maxima_cantidad_vendida_tienda1} unidades')
print(f'El producto más vendido en la tienda 2 es {ventas_producto_tienda2} con {maxima_cantidad_vendida_tienda2} unidades')
print(f'El producto más vendido en la tienda 3 es {ventas_producto_tienda3} con {maxima_cantidad_vendida_tienda3} unidades')
print(f'El producto más vendido en la tienda 4 es {ventas_producto_tienda4} con {maxima_cantidad_vendida_tienda4} unidades')
```

```python
# Producto menos vendido (nombre)
ventas_producto_tienda1_min = tienda['Producto'].value_counts().idxmin()
ventas_producto_tienda2_min = tienda2['Producto'].value_counts().idxmin()
ventas_producto_tienda3_min = tienda3['Producto'].value_counts().idxmin()
ventas_producto_tienda4_min = tienda4['Producto'].value_counts().idxmin()

# Cantidad minima del producto vendido
minima_cantidad_vendida_tienda1 = tienda['Producto'].value_counts().min()
minima_cantidad_vendida_tienda2 = tienda2['Producto'].value_counts().min()
minima_cantidad_vendida_tienda3 = tienda3['Producto'].value_counts().min()
minima_cantidad_vendida_tienda4 = tienda4['Producto'].value_counts().min()

print(f'El producto menos vendido en la tienda 1 es: {ventas_producto_tienda1_min} con {minima_cantidad_vendida_tienda1} unidades')
print(f'El producto menos vendido en la tienda 2 es: {ventas_producto_tienda2_min} con {minima_cantidad_vendida_tienda2} unidades')
print(f'El producto menos vendido en la tienda 3 es: {ventas_producto_tienda3_min} con {minima_cantidad_vendida_tienda3} unidades')
print(f'El producto menos vendido en la tienda 4 es: {ventas_producto_tienda4_min} con {minima_cantidad_vendida_tienda4} unidades')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# --- Datos de productos más vendidos ---
productos_mas_vendidos_data = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Producto': [
        ventas_producto_tienda1,
        ventas_producto_tienda2,
        ventas_producto_tienda3,
        ventas_producto_tienda4
    ],
    'Cantidad': [
        maxima_cantidad_vendida_tienda1,
        maxima_cantidad_vendida_tienda2,
        maxima_cantidad_vendida_tienda3,
        maxima_cantidad_vendida_tienda4
    ]
}
df_mas_vendidos = pd.DataFrame(productos_mas_vendidos_data)

# --- Datos de productos menos vendidos ---
productos_menos_vendidos_data = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Producto': [
        ventas_producto_tienda1_min,
        ventas_producto_tienda2_min,
        ventas_producto_tienda3_min,
        ventas_producto_tienda4_min
    ],
    'Cantidad': [
        minima_cantidad_vendida_tienda1,
        minima_cantidad_vendida_tienda2,
        minima_cantidad_vendida_tienda3,
        minima_cantidad_vendida_tienda4
    ]
}
df_menos_vendidos = pd.DataFrame(productos_menos_vendidos_data)

# --- Gráfico de productos más vendidos ---
plt.figure(figsize=(14, 7))
sns.barplot(x='Tienda', y='Cantidad', hue='Producto', data=df_mas_vendidos, palette='viridis')
plt.title('Productos Más Vendidos por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Cantidad de Unidades Vendidas')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Producto Más Vendido', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# --- Gráfico de productos menos vendidos ---
plt.figure(figsize=(14, 7))
sns.barplot(x='Tienda', y='Cantidad', hue='Producto', data=df_menos_vendidos, palette='magma')
plt.title('Productos Menos Vendidos por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Cantidad de Unidades Vendidas')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Producto Menos Vendido', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Preparar los datos (ya están ordenados de mayor a menor, para barras horizontales es mejor invertirlos para que el más grande quede arriba)
lugares = costo_lugar_compras.index[::-1] # Invertir el orden para que el mayor costo quede arriba
costos = costo_lugar_compras['Costo de envío'][::-1]

plt.figure(figsize=(12, 8))
plt.barh(lugares, costos, color='skyblue')
plt.xlabel('Costo de Envío ($)')
plt.ylabel('Lugar de Compra')
plt.title('Costo Total de Envío por Lugar de Compra (Gráfico de Barras Horizontales)')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

df_geo = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

df_geo[['Tienda','lat','lon','Costo de envío','Lugar de Compra']].head()
```

```python
conteo_lugar_compra= df_geo['Lugar de Compra'].value_counts()
print(f'Conteo de lugares de compra:\n{conteo_lugar_compra}')
```

```python
plt.figure(figsize=(9, 7))

# Obtener los nombres de las tiendas únicas
unique_tiendas = df_geo['Tienda'].unique()

# Definir una lista de colores (o usar un mapa de colores de matplotlib)
colors = ['green', 'yellow', 'skyblue', 'pink'] # Se corrigió 'sky blue' a 'skyblue'

# Iterar sobre cada tienda para graficar sus puntos
for i, tienda_name in enumerate(unique_tiendas):
    df_subset = df_geo[df_geo['Tienda'] == tienda_name]
    plt.scatter(
        data=df_subset,
        x='lon',
        y='lat',
        s=40,
        label=tienda_name, # Etiqueta para la leyenda
        color=colors[i % len(colors)], # Asignar un color de la lista
        alpha=0.6 # Hacer los puntos semi-transparentes
    )

plt.title('Distribución Geográfica de las Ventas por Tiendas')
plt.xlabel('Longitud (lon)')
plt.ylabel('Latitud (lat)')
plt.legend(title='Tienda')
plt.show()
```

# Análisis del desempeño Geográfico

```python
import plotly.express as px
import pandas as pd

# Agrega los datos para crear 'Ventas_Agrupadas' y  'Num_Transacciones'
df_ventas_agrupadas = df_geo.groupby(['Tienda', 'lat', 'lon']).agg(
    Ventas_Agrupadas=('Precio', 'sum'),
    Num_Transacciones=('Precio', 'count')
).reset_index()

fig_concentracion = px.scatter_mapbox(
    df_ventas_agrupadas,
    lat="lat",
    lon="lon",
    size="Ventas_Agrupadas",
    color="Ventas_Agrupadas",
    color_continuous_scale=px.colors.cyclical.IceFire, # Escala de colores
    hover_name="Tienda", # Muestra el nombre de la tienda al pasar el mouse
    hover_data=['Ventas_Agrupadas', 'Num_Transacciones'], # Información adicional
    zoom=4,
    height=600,
    title="Concentración de Ventas por Ubicación (El tamaño = Ventas)"
)

fig_concentracion.update_layout(mapbox_style="carto-positron")
fig_concentracion.show()
```
